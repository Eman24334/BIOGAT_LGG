```python
# Graph Attention Network Using MultiOmics Data with Top Gene Identification
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import KFold
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score
from torch_geometric.data import Data
from torch_geometric.nn import GATConv
from torch.nn import BatchNorm1d, LeakyReLU
import torch.nn.functional as F
from collections import defaultdict
import matplotlib.pyplot as plt

# Check for GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Load and preprocess data
file_path = '/kaggle/input/scaled-lasso-selected-features/scaled_Lasso_selected_features.csv'
df = pd.read_csv(file_path)
Y = df.iloc[:, -1].copy()
feature_names = df.columns[:-1].tolist()  # Store feature names
df = df.iloc[:, :-1].select_dtypes(include=[np.number])  # Keep only numeric columns

# Standardize and encode
X = StandardScaler().fit_transform(df.values)
Y = LabelEncoder().fit_transform(Y)

# Enhanced GAT Model with Attention Tracking
class GATWithAttention(nn.Module):
    def __init__(self, num_features):
        super(GATWithAttention, self).__init__()
        # Layer configurations
        self.conv1 = GATConv(num_features, 512, heads=8)
        self.conv2 = GATConv(512*8, 256, heads=4)
        self.conv3 = GATConv(256*4, 128, heads=2) 
        self.conv4 = GATConv(128*2, 32, heads=1)
        
        # Normalization and activation
        self.bn1 = BatchNorm1d(512*8)
        self.bn2 = BatchNorm1d(256*4)
        self.bn3 = BatchNorm1d(128*2)
        self.bn4 = BatchNorm1d(32)
        self.relu = LeakyReLU()
        self.dropout = nn.Dropout(0.5)
        
        # Attention storage
        self.attention_weights = []

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        self.attention_weights = []  # Reset attention storage
        
        # Layer 1
        x, att1 = self.conv1(x, edge_index, return_attention_weights=True)
        self.attention_weights.append(att1)
        x = self.relu(self.bn1(x))
        x = self.dropout(x)
        
        # Layer 2
        x, att2 = self.conv2(x, edge_index, return_attention_weights=True)
        self.attention_weights.append(att2)
        x = self.relu(self.bn2(x))
        x = self.dropout(x)
        
        # Layer 3
        x, att3 = self.conv3(x, edge_index, return_attention_weights=True)
        self.attention_weights.append(att3)
        x = self.relu(self.bn3(x))
        x = self.dropout(x)
        
        # Layer 4
        x, att4 = self.conv4(x, edge_index, return_attention_weights=True)
        self.attention_weights.append(att4)
        x = self.relu(self.bn4(x))
        x = self.dropout(x)
        
        return F.log_softmax(x, dim=1)

# Robust Attention Weight Processing
def process_attention_weights(attention_weights, feature_names):
    gene_importance = defaultdict(float)
    
    for edge_index, edge_weights in attention_weights:
        edge_index = edge_index.cpu().numpy()
        edge_weights = edge_weights.cpu().numpy()
        
        # Handle multi-head attention
        if len(edge_weights.shape) > 1:
            edge_weights = np.mean(edge_weights, axis=1)
        
        # Update gene importance scores
        for i in range(edge_index.shape[1]):
            src, dst = edge_index[:, i]
            weight = edge_weights[i]
            
            if src < len(feature_names):
                gene_importance[feature_names[src]] += weight
            if dst < len(feature_names):
                gene_importance[feature_names[dst]] += weight
                
    return gene_importance

# Training and Evaluation
k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=42)
metrics = {'accuracy': [], 'precision': [], 'recall': [], 'f1': []}
all_top_genes = []

for fold_idx, (train_idx, test_idx) in enumerate(kf.split(X)):
    print(f"\nFold {fold_idx+1}/{k}")
    
    # Prepare data
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = Y[train_idx], Y[test_idx]
    
    # Create correlation graphs
    corr_train = np.corrcoef(X_train)
    corr_test = np.corrcoef(X_test)
    threshold = 0.95
    
    # Create edge indices
    edge_index_train = torch.tensor(np.where(corr_train > threshold), dtype=torch.long).to(device)
    edge_index_test = torch.tensor(np.where(corr_test > threshold), dtype=torch.long).to(device)
    
    # Create data objects
    train_data = Data(
        x=torch.FloatTensor(X_train).to(device),
        edge_index=edge_index_train
    )
    test_data = Data(
        x=torch.FloatTensor(X_test).to(device),
        edge_index=edge_index_test
    )
    
    # Initialize model
    model = GATWithAttention(X.shape[1]).to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
    criterion = nn.NLLLoss()
    
    # Training loop
    best_acc = 0
    for epoch in range(100):
        model.train()
        optimizer.zero_grad()
        out = model(train_data)
        loss = criterion(out, torch.LongTensor(y_train).to(device))
        loss.backward()
        optimizer.step()
        
        # Validation
        model.eval()
        with torch.no_grad():
            logits = model(test_data)
            pred = logits.argmax(dim=1)
            acc = accuracy_score(y_test, pred.cpu().numpy())
            
            if acc > best_acc:
                best_acc = acc
                torch.save(model.state_dict(), f'best_model_fold{fold_idx}.pt')
                
        if (epoch+1) % 10 == 0:
            print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}, Val Acc: {acc:.4f}")
    
    # Load best model
    model.load_state_dict(torch.load(f'best_model_fold{fold_idx}.pt'))
    
    # Final evaluation
    model.eval()
    with torch.no_grad():
        logits = model(test_data)
        pred = logits.argmax(dim=1)
        
        # Store metrics
        metrics['accuracy'].append(accuracy_score(y_test, pred.cpu().numpy()))
        metrics['precision'].append(precision_score(y_test, pred.cpu().numpy(), average='macro'))
        metrics['recall'].append(recall_score(y_test, pred.cpu().numpy(), average='macro'))
        metrics['f1'].append(f1_score(y_test, pred.cpu().numpy(), average='macro'))
        
        # Process attention weights
        gene_importance = process_attention_weights(model.attention_weights, feature_names)
        sorted_genes = sorted(gene_importance.items(), key=lambda x: x[1], reverse=True)
        top_genes = [gene for gene, score in sorted_genes[:100]]
        all_top_genes.append(top_genes)
        
        print(f"Top 5 genes this fold: {top_genes[:5]}")

# Calculate and display final metrics
print("\nFinal Metrics:")
for metric, values in metrics.items():
    print(f"{metric.capitalize()}: {np.mean(values):.4f} ± {np.std(values):.4f}")

# Generate consensus gene ranking
gene_freq = defaultdict(int)
gene_scores = defaultdict(float)

for fold_genes in all_top_genes:
    for rank, gene in enumerate(fold_genes):
        gene_freq[gene] += 1
        gene_scores[gene] += (100 - rank)  # Higher score for higher ranks

# Create final ranking
final_ranking = sorted(
    [(gene, gene_freq[gene], gene_scores[gene]) for gene in gene_freq],
    key=lambda x: (-x[1], -x[2])  # Sort by frequency then score
)

# Save and display results
top_100_genes = [gene for gene, freq, score in final_ranking[:100]]
print("\nTop 10 Consensus Genes:")
for i, (gene, freq, score) in enumerate(final_ranking[:10]):
    print(f"{i+1}. {gene} (in {freq} folds, score: {score:.1f})")

# Save to file
pd.DataFrame(final_ranking[:100], columns=['Gene', 'Frequency', 'Score'])\
    .to_csv('top_100_genes.csv', index=False)

# Visualization
plt.figure(figsize=(10, 6))
plt.barh([x[0] for x in final_ranking[:20]][::-1], 
         [x[1] for x in final_ranking[:20]][::-1])
plt.xlabel('Frequency in Top 100 Across Folds')
plt.title('Top 20 Most Important Genes')
plt.tight_layout()
plt.savefig('top_genes.png', dpi=300)
plt.show()
```

```python
# Graph Attention Network Using MultiOmics Data with Top Gene Identification
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import KFold
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score
from torch_geometric.data import Data
from torch_geometric.nn import GATConv
from torch.nn import BatchNorm1d, LeakyReLU
import torch.nn.functional as F
from collections import defaultdict
import matplotlib.pyplot as plt

# Check for GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Load and preprocess data
file_path = '/kaggle/input/scaled-lasso-selected-features/scaled_Lasso_selected_features.csv'
df = pd.read_csv(file_path)
Y = df.iloc[:, -1].copy()
feature_names = df.columns[:-1].tolist()  # Store feature names
df = df.iloc[:, :-1].select_dtypes(include=[np.number])  # Keep only numeric columns

# Standardize and encode
X = StandardScaler().fit_transform(df.values)
Y = LabelEncoder().fit_transform(Y)

# Enhanced GAT Model with Attention Tracking
class GATWithAttention(nn.Module):
    def __init__(self, num_features):
        super(GATWithAttention, self).__init__()
        # Layer configurations
        self.conv1 = GATConv(num_features, 512, heads=8)
        self.conv2 = GATConv(512*8, 256, heads=4)
        self.conv3 = GATConv(256*4, 128, heads=2) 
        self.conv4 = GATConv(128*2, 32, heads=1)
        
        # Normalization and activation
        self.bn1 = BatchNorm1d(512*8)
        self.bn2 = BatchNorm1d(256*4)
        self.bn3 = BatchNorm1d(128*2)
        self.bn4 = BatchNorm1d(32)
        self.relu = LeakyReLU()
        self.dropout = nn.Dropout(0.5)
        
        # Attention storage
        self.attention_weights = []

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        self.attention_weights = []  # Reset attention storage
        
        # Layer 1
        x, att1 = self.conv1(x, edge_index, return_attention_weights=True)
        self.attention_weights.append(att1)
        x = self.relu(self.bn1(x))
        x = self.dropout(x)
        
        # Layer 2
        x, att2 = self.conv2(x, edge_index, return_attention_weights=True)
        self.attention_weights.append(att2)
        x = self.relu(self.bn2(x))
        x = self.dropout(x)
        
        # Layer 3
        x, att3 = self.conv3(x, edge_index, return_attention_weights=True)
        self.attention_weights.append(att3)
        x = self.relu(self.bn3(x))
        x = self.dropout(x)
        
        # Layer 4
        x, att4 = self.conv4(x, edge_index, return_attention_weights=True)
        self.attention_weights.append(att4)
        x = self.relu(self.bn4(x))
        x = self.dropout(x)
        
        return F.log_softmax(x, dim=1)

# Robust Attention Weight Processing
def process_attention_weights(attention_weights, feature_names):
    gene_importance = defaultdict(float)
    
    for edge_index, edge_weights in attention_weights:
        edge_index = edge_index.cpu().numpy()
        edge_weights = edge_weights.cpu().numpy()
        
        # Handle multi-head attention
        if len(edge_weights.shape) > 1:
            edge_weights = np.mean(edge_weights, axis=1)
        
        # Update gene importance scores
        for i in range(edge_index.shape[1]):
            src, dst = edge_index[:, i]
            weight = edge_weights[i]
            
            if src < len(feature_names):
                gene_importance[feature_names[src]] += weight
            if dst < len(feature_names):
                gene_importance[feature_names[dst]] += weight
                
    return gene_importance

# Training and Evaluation
k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=42)
metrics = {'accuracy': [], 'precision': [], 'recall': [], 'f1': []}
all_top_genes = []

for fold_idx, (train_idx, test_idx) in enumerate(kf.split(X)):
    print(f"\nFold {fold_idx+1}/{k}")
    
    # Prepare data
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = Y[train_idx], Y[test_idx]
    
    # Create correlation graphs
    corr_train = np.corrcoef(X_train)
    corr_test = np.corrcoef(X_test)
    threshold = 0.95
    
    # Create edge indices
    edge_index_train = torch.tensor(np.where(corr_train > threshold), dtype=torch.long).to(device)
    edge_index_test = torch.tensor(np.where(corr_test > threshold), dtype=torch.long).to(device)
    
    # Create data objects
    train_data = Data(
        x=torch.FloatTensor(X_train).to(device),
        edge_index=edge_index_train
    )
    test_data = Data(
        x=torch.FloatTensor(X_test).to(device),
        edge_index=edge_index_test
    )
    
    # Initialize model
    model = GATWithAttention(X.shape[1]).to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
    criterion = nn.NLLLoss()
    
    # Training loop
    best_acc = 0
    for epoch in range(100):
        model.train()
        optimizer.zero_grad()
        out = model(train_data)
        loss = criterion(out, torch.LongTensor(y_train).to(device))
        loss.backward()
        optimizer.step()
        
        # Validation
        model.eval()
        with torch.no_grad():
            logits = model(test_data)
            pred = logits.argmax(dim=1)
            acc = accuracy_score(y_test, pred.cpu().numpy())
            
            if acc > best_acc:
                best_acc = acc
                torch.save(model.state_dict(), f'best_model_fold{fold_idx}.pt')
                
        if (epoch+1) % 10 == 0:
            print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}, Val Acc: {acc:.4f}")
    
    # Load best model
    model.load_state_dict(torch.load(f'best_model_fold{fold_idx}.pt'))
    
    # Final evaluation
    model.eval()
    with torch.no_grad():
        logits = model(test_data)
        pred = logits.argmax(dim=1)
        
        # Store metrics
        metrics['accuracy'].append(accuracy_score(y_test, pred.cpu().numpy()))
        metrics['precision'].append(precision_score(y_test, pred.cpu().numpy(), average='macro'))
        metrics['recall'].append(recall_score(y_test, pred.cpu().numpy(), average='macro'))
        metrics['f1'].append(f1_score(y_test, pred.cpu().numpy(), average='macro'))
        
        # Process attention weights
        gene_importance = process_attention_weights(model.attention_weights, feature_names)
        sorted_genes = sorted(gene_importance.items(), key=lambda x: x[1], reverse=True)
        top_genes = [gene for gene, score in sorted_genes[:100]]
        all_top_genes.append(top_genes)
        
        print(f"Top 5 genes this fold: {top_genes[:5]}")

# Calculate and display final metrics
print("\nFinal Metrics:")
for metric, values in metrics.items():
    print(f"{metric.capitalize()}: {np.mean(values):.4f} ± {np.std(values):.4f}")

# Generate consensus gene ranking
gene_freq = defaultdict(int)
gene_scores = defaultdict(float)

for fold_genes in all_top_genes:
    for rank, gene in enumerate(fold_genes):
        gene_freq[gene] += 1
        gene_scores[gene] += (100 - rank)  # Higher score for higher ranks

# Create final ranking
final_ranking = sorted(
    [(gene, gene_freq[gene], gene_scores[gene]) for gene in gene_freq],
    key=lambda x: (-x[1], -x[2])  # Sort by frequency then score
)

# Save and display results
top_100_genes = [gene for gene, freq, score in final_ranking[:100]]
print("\nTop 10 Consensus Genes:")
for i, (gene, freq, score) in enumerate(final_ranking[:10]):
    print(f"{i+1}. {gene} (in {freq} folds, score: {score:.1f})")

# Save to file
pd.DataFrame(final_ranking[:100], columns=['Gene', 'Frequency', 'Score'])\
    .to_csv('top_100_genes.csv', index=False)

# Visualization
plt.figure(figsize=(10, 6))
plt.barh([x[0] for x in final_ranking[:20]][::-1], 
         [x[1] for x in final_ranking[:20]][::-1])
plt.xlabel('Frequency in Top 100 Across Folds')
plt.title('Top 20 Most Important Genes')
plt.tight_layout()
plt.savefig('top_genes.png', dpi=300)
plt.show()

import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# --------- 1. رسم Attention Map لكل الطبقات ---------
def plot_attention_maps(attention_weights, feature_names=None):
    for i, (edge_index, edge_weight) in enumerate(attention_weights):
        print(f"Plotting Layer {i+1} Attention Map...")
        
        edge_index = edge_index.cpu().numpy()
        edge_weight = edge_weight.cpu().numpy()

        # Average across heads if multi-head
        if len(edge_weight.shape) > 1:
            edge_weight = np.mean(edge_weight, axis=1)

        G = nx.Graph()
        for j in range(edge_index.shape[1]):
            src = edge_index[0, j]
            dst = edge_index[1, j]
            weight = edge_weight[j]

            src_name = feature_names[src] if feature_names and src < len(feature_names) else src
            dst_name = feature_names[dst] if feature_names and dst < len(feature_names) else dst

            G.add_edge(src_name, dst_name, weight=weight)

        pos = nx.spring_layout(G, seed=42)
        edges, weights = zip(*nx.get_edge_attributes(G, 'weight').items())

        plt.figure(figsize=(10, 8))
        nx.draw(G, pos, edges=edges, edge_color=weights, width=2.0, edge_cmap=plt.cm.viridis,
                node_size=50, with_labels=False)
        plt.title(f"Attention Map - Layer {i+1}")
        plt.tight_layout()
        plt.savefig(f"attention_map_layer{i+1}.png", dpi=300)
        plt.show()

# --------- 2. رسم Attention Map لأهم الجينات فقط ---------
def plot_top_gene_attention(attention_weights, feature_names, top_genes, layer_index=0):
    edge_index, edge_weight = attention_weights[layer_index]
    edge_index = edge_index.cpu().numpy()
    edge_weight = edge_weight.cpu().numpy()

    if len(edge_weight.shape) > 1:
        edge_weight = np.mean(edge_weight, axis=1)

    index_to_gene = {i: name for i, name in enumerate(feature_names)}
    G = nx.Graph()
    for i in range(edge_index.shape[1]):
        src = index_to_gene.get(edge_index[0, i])
        dst = index_to_gene.get(edge_index[1, i])
        weight = edge_weight[i]

        if src in top_genes and dst in top_genes:
            G.add_edge(src, dst, weight=weight)

    if G.number_of_edges() == 0:
        print("No edges found among top genes.")
        return

    pos = nx.spring_layout(G, seed=42)
    edges, weights = zip(*nx.get_edge_attributes(G, 'weight').items())

    plt.figure(figsize=(10, 8))
    nx.draw(G, pos, edges=edges, edge_color=weights, width=2.0,
            edge_cmap=plt.cm.viridis, node_size=600, node_color='lightblue',
            with_labels=True, font_size=8)
    plt.title(f"Top Gene Attention Map - Layer {layer_index+1}")
    plt.tight_layout()
    plt.savefig(f"top_gene_attention_map_layer{layer_index+1}.png", dpi=300)
    plt.show()
```
